# Hypothesis and Findings

## Hypothesis

There is a relationship between menstrual cycle phases and physical and emotional states such as mood, cramps, and energy levels.

## Findings

The analysis reveals clear and consistent patterns across different phases of the menstrual cycle.

Mood is highest during the non-menstrual (other) phase and drops significantly during the premenstrual phase. This indicates that emotional well-being is negatively affected as the cycle approaches menstruation.

Cramps are highest during the menstrual phase, while they are nearly absent during the rest of the cycle. A slight increase is observed in the premenstrual phase, suggesting early physical discomfort before menstruation begins.

Energy levels follow a similar pattern to mood. Energy is highest during the non-menstrual phase and decreases during both menstrual and premenstrual phases. The lowest energy levels are observed just before menstruation.

Overall, these findings strongly support the hypothesis that menstrual cycle phases influence both physical symptoms and emotional well-being.

## Conclusion

This project demonstrates that menstrual cycle phases have a measurable impact on daily well-being. Both emotional (mood) and physical (cramps, energy) indicators vary systematically throughout the cycle. These findings highlight the importance of considering biological cycles when analyzing human behavior and health-related data.
